```python
import pandas as pd
from pandas import Series, DataFrame

obj = pd.Series([4, 7, -5, 3]) #Series: 1차원 배열 자료구조 ,np.array와 유사
print(obj)
#idx  num
# 0    4
# 1    7
# 2   -5
# 3    3
# dtype: int64
print(obj.array)   #배열 객체
# <NumpyExtensionArray>
# [4, 7, -5, 3]
# Length: 4, dtype: int64
print(obj.index)   #색인 객체
# RangeIndex(start=0, stop=4, step=1)

obj2 = pd.Series([4, 7, -5, 3], index = ["d", "b", "a", "c"]) #index 별도 지정
print(obj2)
# d    4
# b    7
# a   -5
# c    3
# dtype: int64
print(obj2.index)
# Index(['d', 'b', 'a', 'c'], dtype='object')
print(obj2["d"])
# 4

#Serie는 "항상" index 와 값 연결이 유지됨!
print(obj2[["d", "a"]])
# d    4
# a   -5
# dtype: int64
print(obj2[obj2 > 0])
# d    4
# b    7
# c    3
# dtype: int64
print(obj2 * 2)
# d     8
# b    14
# a   -10
# c     6
# dtype: int64
import numpy as np
print(np.exp(obj2))
# d      54.598150
# b    1096.633158
# a       0.006738
# c      20.085537
# dtype: float64

print("e" in obj2)
#False


#파이선 딕셔너리 <-> Series
dic_data = {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000}

obj3 = pd.Series(dic_data)  #dic->Series
print(obj3)
# Ohio     35000
# Texas    71000
# Oregon   16000
# Utah      5000
# dtype: int64
print(obj3.to_dict())  #Series->dict
# {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}

states = ["California", "Ohio", "Oregon", "Texas"]
obj4 = pd.Series(dic_data, index=states) #index 명시
print(obj4)
# California    NaN
# Ohio       35000.0
# Oregon     16000.0
# Texas      71000.0
# dtype: float64

# California에 대한 값을 찾을 수 없기 때문에 이 값은 NaN(not a number)으로 표시
# Utah는 states에 포함되어 있지 않으므로, 실행 결과에는 빠지게 됨

#isna, notna: 누락된 데이터 이다/아니다(True/False)
print(pd.isna(obj4))  #또는 obj4.isna()
# California     True
# Ohio          False
# Oregon        False
# Texas         False
# dtype: bool

print(pd.notna(obj4))
# California    False
# Ohio           True
# Oregon         True
# Texas          True
# dtype: bool

# Series 객체 간 연산 (더하기)
print(obj3 + obj4)
# California     NaN
# Ohio        70000.0
# Oregon      32000.0
# Texas      142000.0
# Utah           NaN
# dtype: float64


# Series와 인덱스에 이름 부여하기
obj4.name = "population"
obj4.index.name = "state"
print(obj4)
# state
# California     NaN
# Ohio       35000.0
# Oregon     16000.0
# Texas      71000.0
# Name: population, dtype: float64



# 대입으로 Series의 색인을 변경.
obj5 = pd.Series([4, 7, -5, 3])  # 기본 숫자 인덱스를 갖는 Series 생성
print(obj5)
# 출력:
# 0    4
# 1    7
# 2   -5
# 3    3
# dtype: int64

# 인덱스 변경하기
obj5.index = ["Bob", "Steve", "Jeff", "Ryan"]
print(obj5)
# 출력:
# Bob      4
# Steve    7
# Jeff    -5
# Ryan     3
# dtype: int64
```
