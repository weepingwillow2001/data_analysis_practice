```python
# pandas 라이브러리 import
import pandas as pd

# 기본 CSV 파일 읽기
df = pd.read_csv("ex1.csv")
df
```

```python
# 헤더가 없는 CSV 파일 읽기
# 헤더: 파일의 "첫" 번째 행으로, 각 "열(column)의 이름"을 담고 있음
pd.read_csv("ex2.csv", header=None) # header = None 옵션을 사용하면 자동으로 "열 이름"을 0부터 "숫자"로 생성
```

```python
# 열 이름(names) 지정하기
pd.read_csv("ex2.csv", names=["a", "b", "c", "d", "message"])
```

```python
# message "열"을 "인덱스"로 사용하기: index_col
headers = ["a", "b", "c", "d", "message"]
pd.read_csv("ex2.csv", names=headers, index_col="message")
```

```python
# "계층"적 "인덱스" 만들기: index_col = [,]
parsed = pd.read_csv("csv_mindex.csv", index_col=["key1", "key2"])
parsed
```

```python
# 구분자(sep)로 파일 읽기
result = pd.read_csv("ex3.txt", sep="\t") # tab으로 구분된 텍스트 파일
result
```

```python
# 주석이 있는 파일 특정 "행" 건너뛰기: skiprows
pd.read_csv("ex4.csv", skiprows=[0, 2, 3]) # ex4 파일의 0행 2행 3행 스킵
```

```python
# 기본적으로 pandas는 다음과 같은 문자열들을 NA 값으로 자동 인식합니다:

# 'NA'
# 'N/A'
# 'nan'
# 'null'
# 'NULL'
# ''(빈 문자열)
# '#N/A'

# 누락된 값 처리할 때 "기본" 값은 keep_default_na = True !!
# NA 값으로 자동 인식하여 NaN 출력
```

```python
# 누락된 값 처리하기
result = pd.read_csv("ex5.csv")
result

# 누락된 값 확인하기
pd.isna(result) # 공백값은 True 로 표시됨
```

```python
# 문자열 지정하여 NaN 출력: na_values = []
result = pd.read_csv("ex5.csv", na_values=["NULL"]) # 또는 pd.read_csv("ex5.csv")
# "NULL" 문자열이 존재하지 않으므로 na_values=["NULL"] 은 무의미함!
result # 공백깂에 "NaN" 출력
```

```python
# 기본 NA 값 표현 변경하기
result2 = pd.read_csv("ex5.csv", keep_default_na=False) # 원본 공백값(문자열) "유지"-빈 칸도 빈 문자열 취급!
result2
# 	something	a	b	 c	 d	message
# 0	one	      1	2	 3	 4	NA
# 1	two	      5	6		   8	world
# 2	three	    9	10 11  12	foo

pd.isna(result2)
```

```python
# NA를 NaN으로 처리하기
result3 = pd.read_csv("ex5.csv", keep_default_na=False, na_values=["NA"])
# "NA" 문자열을 지정하여 NaN 값으로 변환! 나머지는 "유지"
result3

#   something	a	b	 c	 d	message
# 0	one	      1	2	 3	 4	NaN
# 1	two	      5	6		   8	world
# 2	three	    9	10 11	12	foo
result3.isna()
```

```python
# 여러 다른 NA 문자열 처리하기
sentinels = {"message": ["foo", "NA"], "something": ["two"]}
pd.read_csv("ex5.csv", na_values=sentinels, keep_default_na=False)
# sentinels 객체를 지정하여 NaN 값으로 변환! 나머진 "유지"
```
