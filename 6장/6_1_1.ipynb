```python
# 청크(chunk): "대용량" 파일을 처리할 때 전체 파일을 한 번에 메모리에 로드하지 않고,
# 일정 크기의 부분(청크)으로 "나누어 순차적"으로 처리하는 방법입니다.

# 파일을 청크(chunk) 단위로 나누어 읽기: chunksize
chunker = pd.read_csv("ex6.csv", chunksize=1000)  # "파일을 1000"행"씩 나누어 읽겠다"
type(chunker)
# pandas.io.parsers.readers.TextFileReader
```

```python
# 청크 단위로 처리하는 예시 (key 열의 값 집계)
chunker = pd.read_csv("ex6.csv", chunksize=1000)
tot = pd.Series([], dtype='int64')
for piece in chunker:
    tot = tot.add(piece["key"].value_counts(), fill_value=0)
tot = tot.sort_values(ascending=False)

# 결과 확인
tot[:10]
```
