```python
import pandas as pd
import numpy as np
# 랜덤 데이터 생성
frame = pd.DataFrame({"a": np.random.standard_normal(100)})
frame
```

```python
# HDF5 파일 저장소 생성: HDFStore
store = pd.HDFStore("mydata.h5")  # "mydata.h5"라는 HDF5 파일이 생성됩니다
```

```python
# 데이터 저장
store["obj1"] = frame   #  DataFrame인 frame을 HDF5 저장소 내에 "obj1"이라는 이름(또는 키)으로 저장
store["obj1_col"] = frame["a"]  # frame의 "a"열을 HDF5 저장소 내에 "obj1_col"이라는 이름(또는 키)으로 저장

# 저장소 정보 확인
store
```

```python
# 저장된 데이터 접근
store["obj1"]
```

```python
# 테이블 형식으로 저장: format
store.put("obj2", frame, format="table")
```

```python
# 조건으로 데이터 선택: where
store.select("obj2", where="index >= 10 and index <= 15")
# 저장소 닫기
store.close()
```

```python
import tables
tables.file._open_files.close_all()
# 빠른 저장 및 읽기: to_hdf/read_hdf
frame.to_hdf("mydata.h5", "obj3", format="table", mode = "w")

pd.read_hdf("mydata.h5", "obj3", where="index < 5")
```
