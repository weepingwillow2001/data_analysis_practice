```python
# GitHub API 요청
import requests
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
resp = requests.get(url)
resp.raise_for_status()
# HTTP 응답의 상태 코드를 확인하는 메서드입니다. 만약 상태 코드가 4XX(클라이언트 오류) 또는 5XX(서버 오류)인 경우, 예외를 발생시킴
```

```python
# JSON 데이터 처리
data = resp.json()
data[0]["title"]
# data 리스트의 첫 번째 항목(0번째 인덱스)에서 "title" 키의 값을 가져옴
```

```python
import pandas as pd
# 데이터프레임으로 변환
issues = pd.DataFrame(data, columns=["number", "title", "labels", "state"])
issues
```
