```python
# SQLite 데이터베이스 생성
import pandas as pd
import sqlite3
query = """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
 c REAL,        d INTEGER
);"""
con = sqlite3.connect("mydata.sqlite")
con.execute(query)  # execute(): SQL 명령을 "실행"합니다
con.commit()  # commit(): 실행된 변경사항을 실제 데이터베이스 파일에 영구적으로 "저장"합니다.

```

```python
# 데이터 삽입
data = [("Atlanta", "Georgia", 1.25, 6),
        ("Tallahassee", "Florida", 2.6, 3),
        ("Sacramento", "California", 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"  # SQL 명령문(statement)
# "test" 테이블에 값을 삽입하라는 SQL 문입니다. 여기서 물음표(?)는 나중에 채워질 매개변수의 위치를 나타내는 플레이스홀더(placeholder)입니다.
con.executemany(stmt, data)
# data 리스트의 각 튜플을 물음표 위치에 대입하여 실행합니다. 이렇게 하면 동일한 구조의 SQL 명령을 여러 데이터 세트에 대해 효율적으로 실행할 수 있습니다.

con.commit()
```

```python
# 데이터 조회
cursor = con.execute("SELECT * FROM test") # SQL 명령문
# 이 명령문은 "test" 테이블에서 "모든" 열과 "모든" 행을 가져오라는 의미입니다.("*")
rows = cursor.fetchall()  # fetchall: 모든 "행" 데이터 가져오기
rows

```

```python
# 열 정보 확인 및 데이터프레임 변환
cursor.description # descripton으로 "열" 정보 확인
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])  # 리스트 컴프리헨션을 사용하여 각 "열의 이름"을 추출

# cursor.description의 각 요소 x에서
# 첫 번째 요소 "x[0](열 이름)"만 가져와서
# 새 리스트를 만듭니다.
```

```python
# SQLAlchemy 사용

import sqlalchemy as sqla
db = sqla.create_engine("sqlite:///mydata.sqlite")
pd.read_sql("SELECT * FROM test", db)
```
