```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import sys
# JSON 문자열

obj = """
{
"name": "Wes",
"cities_lived": ["Akron", "Nashville", "New York", "San Francisco"],
"pet": null,
"siblings": [{"name": "Scott", "age": 34, "hobbies": ["guitars", "soccer"]},
             {"name": "Katie", "age": 42, "hobbies": ["diving", "art"]}]
}
"""
```

```python
import json

# JSON 문자열 -> "파이썬 객체"로 변환: loads
python = json.loads(obj)

# 결과 확인
python
```

```python
# "파이썬 객체" -> JSON 문자열로 변환: dumps
asjson = json.dumps(python)

asjson
```

```python
# siblings 부분을 DataFrame으로 변환 - 파이선 객체를 이용
siblings = pd.DataFrame(python["siblings"], columns=["name", "age"]) # 열 지정
siblings
```

```python
# JSON 파일 내용 확인
!cat example.json

# pandas로 JSON 파일 읽기: read_jason -> "데이터프레임 생성"됨!
df = pd.read_json("example.json")

# 결과 확인
df
```

```python
# DataFrame을 "JSON으로" 변환 (기본 형식): to_json
df.to_json(sys.stdout) # standard output
```

```python
# DataFrame을 JSON으로 변환
df.to_json(sys.stdout, orient="records")  # records 형식(orient)으로 변환

# 각 "행"이 "독립"된 JSON 객체로 표현되어 직관적입니다
# 많은 웹 API에서 사용하는 표준 형식과 일치합니다
# 데이터를 "행" 단위로 처리하기 쉽습니다
```
