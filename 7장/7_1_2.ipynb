```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

df = pd.DataFrame(np.random.standard_normal((7, 3)))
df.iloc[:4, 1] = np.nan  # 처음 4행의 두 번째 열을 NaN으로 설정
df.iloc[:2, 2] = np.nan  # 처음 2행의 세 번째 열을 NaN으로 설정
df
```

```python
# 누락된 값을 제거하지 않고(완전적으로 다른 데이터도 함께 버리지 가능성이 있다) 데이터 상의 '구멍'을 어떻게 메우고 싶다면 fillna 메서드를 사용한다. fillna 메서드에 채워 넣고 싶은 값을 넘겨주면 된다.
df.fillna(0)
```

```python
# fillna에 딕셔너리 값을 넘기면 각 열마다 다른 값이 채워진다.
df.fillna({1: 0.5, 2: 0})
```

```python
# 재샘플링에서 사용 가능한 보간(interpolation) 메서드(표 5-3)는 fillna 메서드에서도 사용 가능하다.
df = pd.DataFrame(np.random.standard_normal((6, 3)))
df.iloc[2, 1] = np.nan
df.iloc[4, 2] = np.nan
df
```

```python
df.fillna(method="ffill") # forward-fill: 결측치를 "앞"에 값으로 채움("열" 기준!)
```

```python
df.fillna(method="ffill", limit=2)
# limit=2는 연속된 NaN 값  "2개 까지만" 채움
# 3개 이상의 연속된 NaN이 있으면 처음 2개만 채우고 나머지는 NaN으로 그대로 둡니다.
```

```python
# fillna를 이용해서 배우 다양한 일을 할 수 있는데 예를 들면 평균값이나 중간값을 넘겨서 데이터를 채울 수도 있다.
data = pd.Series([1., np.nan, 3.5, np.nan, 7])
data.fillna(data.mean())  # (1+3.5+7)/3 = 3.83 을 결측치에 대입
```
