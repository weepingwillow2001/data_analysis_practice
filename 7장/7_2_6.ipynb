```python
# 필요한 모듈 임포트
import numpy as np
import pandas as pd


# 1000행 4열의 표준정규분포를 따르는 랜덤 데이터로 데이터프레임을 생성
data = pd.DataFrame(np.random.standard_normal((1000, 4)))
data
```

```python
# describe()를 통해 각 열의 기본 통계량(개수, 평균, 표준편차, 최소값, 백분위수, 최대값)을 계산
data.describe()
```

```python
# 특정 조건에 맞는 데이터 필터링

# 데이터프레임의 3번째 열(인덱스 2)을 선택
col = data[2]
# 해당 열에서 절댓값이 3보다 큰 값들만 필터링
col[col.abs() > 3]
```

```python
# 전체 데이터프레임에서 조건에 맞는 행 필터링

# 전체 데이터프레임에서 절댓값이 3보다 큰 값을 필터링
# any(axis="columns")를 사용하여 각 "행"에서 조건을 만족하는 값이 하나라도 있는지 확인: any -> 불리언 배열 반환
data[data.abs() > 3].any(axis="columns")

# any는 선택, 제거 카테고리가 아니므로(불리언 배열로 "변환")!!
```

```python
# 조건에 따른 데이터 변환

# 절댓값이 3을 초과하는 값들을 찾아서 np.sign(data)를 사용해 원래 값의 부호(1 또는 -1)를 유지하면서 절댓값을 3으로 제한
data[data.abs() > 3] = np.sign(data) * 3
# np.sign(data): 데이터의 부호를 나타내는 배열을 생성합니다.

# 양수인 경우 1
# 음수인 경우 -1
# 0인 경우 0

# 값이 4.7인 경우 → 3.0으로 변경
# 값이 -5.2인 경우 → -3.0으로 변경
np.sign(data) * 3
```

```python
# np.sign(data)는 data가 양수인지 음수인지에 따라 1이나 -1이 담긴 배열을 반환
np.sign(data).head()
```
