```python
import numpy as np
import pandas as pd
```

```python
# 무작위 샘플링과 재배열: random.permutation(뒤섞기)

# 5행 7열의 데이터프레임 생성 (0부터 34까지의 값)
df = pd.DataFrame(np.arange(5 * 7).reshape((5, 7)))

df
```

```python
# "행"의 무작위 재배열(permutaion-뒤섞기)

# 0부터 4까지의 숫자를 무작위로 재배열한 인덱스를 생성
sampler = np.random.permutation(5)

sampler
```

```python
# take 메서드를 사용해 "해당 인덱스 순서대로" "행"을 재배열
df.take(sampler)
# "iloc"을 사용해도 동일한 결과
df.iloc[sampler]
```

```python
# "열"의 무작위 재배열

# 0부터 6까지의 숫자를 무작위로 재배열한 인덱스를 생성
column_sampler = np.random.permutation(7)

column_sampler
```

```python
# 해당 인덱스 순서대로 "열"을 재배열
df.take(column_sampler, axis="columns")

# take 메서드는 "선택/제거" 카테고리에 속하기 때문에 axis=columns 또는 axis=1을 사용!!!
```

```python
# 무작위 샘플링: sample

# 데이터프레임에서 3개의 "행"을 무작위로 추출
df.sample(n=3)
```

```python
# 데이터프레임에서 3개의 "열"을 무작위로 추출
df.sample(n=3, axis = 'columns')

# sample 메서드는 "선택/제거" 카테고리에 속하기 때문에 axis=columns 또는 axis=1을 사용!!!
```

```python
# "중복 허용" 샘플링: replace=True
# "5개" 값을 가진 Series에서 중복을 허용하여 "10개"의 샘플을 무작위로 추출
choices = pd.Series([5, 7, -1, 6, 4])
choices.sample(n=10, replace=True)
```
